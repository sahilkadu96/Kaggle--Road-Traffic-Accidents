```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('RTA Dataset.csv')
df
```

```python
def getHour(timing):
    output = []
    for i in range(len(timing)):
        hour = timing[i].split(":")[0]
        output.append(int(hour))
    return output
```

```python
df['TimeHr'] = getHour(df['Time'])
df.head()
```

```python
acc = df['TimeHr'].value_counts().sort_index()
plt.plot(acc.index, acc.values, marker = 'o')
plt.xlabel('TimeHr')
plt.ylabel('Accidents')
plt.show()
```
