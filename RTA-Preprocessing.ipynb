```python
import pandas as pd
```

```python
df = pd.read_csv('RTA Dataset.csv')
df
```

# Preprocessing time

```python
def getHour(timing):
    output = []
    for i in range(len(timing)):
        hour = timing[i].split(":")[0]
        output.append(int(hour))
    return output
```

```python
df['TimeHr'] = getHour(df['Time'])
df.head()
```

```python
df.columns
```

```python
#dropping columns which are not needed
df.drop(['Time', 'Educational_level', 'Vehicle_driver_relation', 'Owner_of_vehicle', 'Service_year_of_vehicle',
        'Work_of_casuality', 'Fitness_of_casuality', 'Defect_of_vehicle'], axis = 1, inplace = True)
df.head()
```

# Filing Missing values

```python
df.isnull().sum()
```

```python
df['Driving_experience'].fillna(df['Driving_experience'].mode()[0], inplace = True)
df['Type_of_vehicle'].fillna(df['Type_of_vehicle'].mode()[0], inplace = True)
df['Area_accident_occured'].fillna(df['Area_accident_occured'].mode()[0], inplace = True)
df['Type_of_vehicle'].fillna(df['Type_of_vehicle'].mode()[0], inplace = True)
df['Lanes_or_Medians'].fillna(df['Lanes_or_Medians'].mode()[0], inplace = True)
df['Road_allignment'].fillna(df['Road_allignment'].mode()[0], inplace = True)
df['Types_of_Junction'].fillna(df['Types_of_Junction'].mode()[0], inplace = True)
df['Road_surface_type'].fillna(df['Road_surface_type'].mode()[0], inplace = True)
df['Type_of_collision'].fillna(df['Type_of_collision'].mode()[0], inplace = True)
df['Vehicle_movement'].fillna(df['Vehicle_movement'].mode()[0], inplace = True)
```

```python
df_Lanes_or_Medians = pd.get_dummies(df['Lanes_or_Medians'], prefix = 'Lanes_or_Medians', drop_first = True)
df_Lanes_or_Medians.head()
```

# Combining all dummy columns

```python
df_dummies = pd.concat([df_Age_band_of_driver, df_Driving_experience, df_Type_of_vehicle, df_Area_accident_occured,
                       df_Lanes_or_Medians, df_Road_allignment, df_Types_of_Junction, df_Road_surface_type,
                       df_Road_surface_conditions, df_Light_conditions, df_Weather_conditions, df_Type_of_collision, df_Vehicle_movement,
                       df_Casualty_class, df_Age_band_of_casualty, df_Pedestrian_movement, df_Cause_of_accident], axis = 1)

df_dummies.head()
```

```python
df_final = pd.concat([df, df_dummies], axis = 1)
df_final.head()
```

```python
df_final.drop(['Age_band_of_driver', 'Driving_experience', 'Type_of_vehicle', 'Area_accident_occured', 'Lanes_or_Medians',
               'Road_allignment', 'Types_of_Junction', 'Road_surface_type', 'Road_surface_conditions', 'Light_conditions',
               'Weather_conditions', 'Type_of_collision', 'Vehicle_movement', 'Casualty_class', 'Age_band_of_casualty', 
               'Pedestrian_movement', 'Cause_of_accident'
              ], axis = 1, inplace = True)
df_final.head()
```
